```python
import matplotlib.pyplot as plt

x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [23, 24, 25, 26, 27, 28, 29, 30, 31]
y2 = [22.96, 23.94, 24.91, 25.89, 26.87, 28.04, 29.02, 30.00, 30.98]
value = []
average = float(0)
plt.plot(x, y1, label="sample temperature")
plt.plot(x, y2, label="Measured temperature value")
plt.plot()

plt.xlabel("Order of experimental values")
plt.ylabel("Experimental value")
plt.title("Comparison between the sample temperature and measured temperature values")
plt.legend()
plt.show()
for i in range(10):
  A = (((y1[i-1])-(y2[i-1]))/(((y1[i-1])+(y2[i-1]))/2))
  value.append(A)
  #print(A)
  average += A
  #print(average)
average /= 10
print("ค่าความแตกต่างคิดเป็น(%)อยู่ที่ "+'%.4f'%(average)+"%")
```
